```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data_frame = pd.read_excel('website_traffic.xls')
data_frame.head()
```

```python
website = 'www.google.com.af'
lis = website.split('.')
if lis[-1] != 'com':
    print(lis[1] + '.' + lis[-1])
```

```python
def extract_wesite_name(url:str) -> pd.Series:
    lis = url.split('.')
    if lis[-1] != 'com':
        return lis[1] + '.' + lis[-1]
    else:
        return url.split('.')[1]

data_frame['Website'] = data_frame['Website'].apply(extract_wesite_name)
```

```python
col_name = 'Country_Rank'
(' ').join([i.lower() for i in col_name.split('_')])
```

```python
column_name = []
for col in data_frame.columns:
    column_name.append((' ').join([i.capitalize() for i in col.split('_')]))
data_frame.columns = column_name
```

```python
import numpy as np
data_frame = data_frame.replace('-', np.nan)
data_frame.isnull().sum().plot(kind='bar')
```

```python
data_frame.head()
```

```python
data_frame.dropna(subset=['Country'], inplace=True)
data_frame.reset_index(drop=True, inplace=True)
data_frame.shape
```

```python
data_frame[data_frame['Location'].isna()].head()
```

```python
def find_imputes(data_frame, null_col:str) -> None:
    """
    find the possible impute values for missing values in each column
    """
    shows = []
    for website in data_frame[data_frame[null_col].isna()]['Website']:
        if website not in shows:
            shows.append(website)
            print(website, ':', data_frame[data_frame['Website']==website][null_col].unique())
            
find_imputes(data_frame, 'Location')
```

```python
def impute_location(location:str, country:str):
    if pd.isnull(location):
        location = country
    else:
        location = location
    return location

data_frame['Location'] = data_frame.apply(lambda x: impute_location(x['Location'], x['Country']), axis=1)
```

```python
data_frame.shape
```

```python
def to_numeric(df, cols):
    for col in cols:
        df[col] = df[col].replace({'K': '*1e3', 'M': '*1e6', np.nan:-1}, regex=True).map(pd.eval).astype(int)
    return df

cols = ('Twitter Mentions', 'Facebook Likes', 'Google Pluses', 'Linkedin Mentions', 'Pinterest Pins', 'Stumbleupon Views')
data_frame = to_numeric(data_frame, cols=cols)
```

```python
data_frame.select_dtypes(include=['object']).columns
```

```python
cols = ('Avg Daily Visitors', 'Avg Daily Pageviews', 'Traffic Rank')
def type_conversion(df, col_tuple, to_type):
    for cols in col_tuple:
        df[cols] = df[cols].str.replace(' ', '').astype(to_type)
    return df

data_frame = type_conversion(data_frame, cols, 'float')
```

```python
def fetch_most_popular_websites(country):
    return data_frame[data_frame.Country==country][['Website', 'Country Rank', 'Facebook Likes',
                                                   'Twitter Mentions', 'Hosted By']][0:5]

print(fetch_most_popular_websites('India'))
```

```python
import squarify
# squarify.plot?
```

```python
def plot_tree(nation):

    country = fetch_most_popular_websites(nation)
    sizes = country['Country Rank'].to_list()
    color = plt.cm.Dark2(len(sizes))
    label = list(zip(country.Website, country['Country Rank'],
                    country['Facebook Likes'], country['Hosted By']))
    
    # treemap plot
    plt.figure(figsize=(12,6))
    squarify.plot(sizes=sizes, 
                  color=color, 
                  label=label, 
                  pad=True)

    # labels
    plt.title('Top 5 Website Rank of {}'.format(nation))
    plt.axis('off')

plot_tree('India')
```

```python
domain_info  = ('Location', 'Registrant', 'Hosted By', 'Registrar', 'Subnetworks')
social_media = ('Facebook Likes', 'Twitter Mentions', 'Google Pluses', 'Linkedin Mentions', 'Pinterest Pins')

def fetch_dimensional_attribute(data_frame:pd.DataFrame, website:str, country:str) -> None:
    """
    fetch dimensional attitubes of websites with countries 
    considering the most repeated values for some features
    """
    print('DOMAIN INFO')
    for i in domain_info:
        try:
            print(i.capitalize(), ':', data_frame[(data_frame['Website']==website) & 
                (data_frame['Country']==country)][i].value_counts().index[0])
        
        except:
            print(i.capitalize(), ':', 'NULL')
    print()
    
    print('SOCIAL MEDIA ACTIVITIES (-1 represents NULL Values)')
    for i in social_media:
        print(i.capitalize(), ':', data_frame[(data_frame['Website']==website) & 
            (data_frame['Country']==country)][i].value_counts().index[0])
    
    print()
    
    print('TRAFFIC ESTIMATES AND TRENDS')
    
    try:
        print('Traffic Rank', ':' ,data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)]['Traffic Rank'].value_counts().index[0])
    except:
        print('Traffic Rank : NULL')
    
    print('Status', ':' ,data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)]['Status'].value_counts().index[0])
    
    print(data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)][['Reach Day', 'Reach Day Percentage']])
    
    print(data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)][['Month Average Daily Reach', 'Month Average Daily Reach Percentage']])
    
    print(data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)][['Daily Pageviews', 'Daily Pageviews Percentage']])
    
    print(data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)][['Month Average Daily Pageviews', 'Month Average Daily Pageviews Percentage']])
    
    print(data_frame[(data_frame['Website']==website) & 
         (data_frame['Country']==country)][['Daily Pageviews Per User', 'Daily Pageviews Per User Percentage']])
    
fetch_dimensional_attribute(data_frame, 'booksecure.net', 'Afghanistan')
```

```python
def find_number_1_website(website:str, rank:int):
    """
    Find where the input website is ranked
    """
    return data_frame[(data_frame['Website'].str.contains(website,
                     case=False, regex=True)) & (data_frame['Country Rank']==rank)]
```
